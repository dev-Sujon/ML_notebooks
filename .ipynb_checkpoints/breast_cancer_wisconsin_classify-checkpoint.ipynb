## Spliting the data

```python
# access dataset columns for train and testing
X=df.drop(['diagnosis'],axis=1)
Y=df['diagnosis']
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
print("x_train shape :",x_train.shape)
print("x_test  shape :",x_test.shape)
print("y_train shape :",y_train.shape)
print("y_test  shape :",y_test.shape)
```

## Machine Learning model

```python
knn=KNeighborsClassifier()
```

### Identify best parameters

```python
# Hyperparameter Tuning
param={
  'n_neighbors':[x for x in range(1,100)],
    'weights':['uniform','distance'],
    'p':[1,2,3],
    'leaf_size':[10,20,30]
}
grid_search=GridSearchCV(knn,param)
grid_search.fit(x_train,y_train)
grid_search.best_params_
```

```python
random_search=RandomizedSearchCV(knn,param)
random_search.fit(x_train,y_train)
random_search.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors = 15,
                         p = 3,
                         weights= 'distance',
                         leaf_size=10,
                         n_jobs=-1
                        )
knn
```

```python
knn.fit(x_train,y_train)
```

## Evaluation

```python
y_predict_test=knn.predict(x_test)
y_predict_train=knn.predict(x_train)
```

### confusion matrix

```python
print('confusion matrix :')
print('test\n',confusion_matrix(y_predict_test,y_test))
print('train\n',confusion_matrix(y_predict_train,y_train))
```

```python
print('classification report :')
print('test\n',classification_report(y_predict_test,y_test))
print('train\n',classification_report(y_predict_train,y_train))
```

### acuuracy score

```python
acc_test=accuracy_score(y_predict_test,y_test)
acc_train=accuracy_score(y_predict_train,y_train)
print(acc_test)
print(acc_train)
```

```python
for i in range(1,42):
    neigh=KNeighborsClassifier(n_neighbors=i)
    neigh.fit(x_train,y_train)
    y_prediction=neigh.predict(x_test)
    accu=accuracy_score(y_prediction,y_test)
    print(f'For {i} n_neighbours accuracy score for test is {accu}')

    y_prediction=neigh.predict(x_train)
    accu=accuracy_score(y_prediction,y_train)
    print(f'For {i} n_neighbours accuracy score for train is {accu}')
    print('*'*100)
```
